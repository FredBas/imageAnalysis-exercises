math.atan2(y, x) returns the arc tan of y/x in radians, between -pi and pi

```python
import math

a = 10
b = 3

theta_rad = math.atan2(b, a)  # angle in radians
theta_deg = math.degrees(theta_rad)  # convert to degrees

print("Theta (radians):", theta_rad)
print("Theta (degrees):", theta_deg)
```

```python
def camera_b_distance(f, g):
    distanceForCCD = 1.0 / (1.0/f - 1.0/g)
    return distanceForCCD
```

### Exercise 3 – Solution

**Given:**
- Thomas’ height: 1.8 m
- Object distance \(g = 5.0\) m
- Focal length \(f = 5\) mm = 0.005 m
- CCD size: 6.4 mm × 4.8 mm (1/2" chip)
- Resolution: 640 × 480 pixels

---

**1. Distance from lens to image (CCD position):**

\[
\frac{1}{g} + \frac{1}{b} = \frac{1}{f} \quad \Rightarrow \quad b \approx 0.0050025\ \text{m} = 5.00\ \text{mm}
\]

✅ The CCD should be placed **~5.00 mm** behind the lens.

---

**2. Thomas’ height on the CCD:**

Magnification:
\[
M = \frac{b}{g} \approx 0.0010005
\]

Image height:
\[
h' = M \cdot h = 0.0010005 \times 1800\ \text{mm} \approx 1.80\ \text{mm}
\]

✅ Thomas appears **~1.80 mm tall** on the CCD.

---

**3. Pixel size:**

\[
\frac{6.4\ \text{mm}}{640} = 0.01\ \text{mm}, \quad \frac{4.8\ \text{mm}}{480} = 0.01\ \text{mm}
\]

✅ Each pixel is **0.01 mm (10 μm)**.

---

**4. Thomas’ height in pixels:**

\[
\frac{1.80}{0.01} \approx 180\ \text{pixels}
\]

✅ Thomas’ image is about **180 pixels tall**.

---

**5. Horizontal field-of-view:**

\[
\text{FOV}_x = 2 \cdot \arctan\left(\frac{6.4}{2 \cdot 5}\right) \approx 65.4^\circ
\]

✅ Horizontal FOV ≈ **65°**

---

**6. Vertical field-of-view:**

\[
\text{FOV}_y = 2 \cdot \arctan\left(\frac{4.8}{2 \cdot 5}\right) \approx 51.2^\circ
\]

✅ Vertical FOV ≈ **51°**

---

### ✅ Final Results
1. Image distance (CCD position): **5.00 mm**
2. Thomas’ image height on CCD: **1.80 mm**
3. Pixel size: **0.01 mm (10 μm)**
4. Thomas’ height in pixels: **180 px**
5. Horizontal FOV: **65°**
6. Vertical FOV: **51°**
